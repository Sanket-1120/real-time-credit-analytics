```python
# In a new cell in your notebook

import sys
sys.path.append('..')

from data_ingestion.sec_edgar_fetcher import fetch_sec_filings
import pandas as pd
from sec_edgar_api import EdgarClient
from dotenv import load_dotenv
import os

# We need to manually run the first few steps of the function
# to see what the raw data looks like.
load_dotenv()
user_name = os.getenv("EDGAR_USER_NAME")
user_email = os.getenv("EDGAR_USER_EMAIL")
edgar_client = EdgarClient(user_agent=f"{user_name} {user_email}")

# Fetch submissions for Apple (CIK: 0000320193)
submissions = edgar_client.get_submissions(cik="0000320193")
filings_df = pd.DataFrame.from_dict(submissions, orient="index")

# --- THIS IS THE IMPORTANT PART ---
print("Available columns are:")
print(filings_df.columns)

print("\nFirst 2 rows of data:")
print(filings_df.head(2))
```
